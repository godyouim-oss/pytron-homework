### ‡∏Ç‡πâ‡∏≠ 9. ‡∏à‡∏á‡πÅ‡∏õ‡∏•‡∏á‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏û‡∏ò‡∏≠‡∏ô (2 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

- a) $\frac{1 + 2 + 3}{4 + 5 + 6}$

- b) ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 27 (Hint: $\sqrt[3]{a} = a^{1/3}$)

- c) $3^{5+\frac{3}{4}}$

````python
# a) 
answer_a = (1 + 2 + 3) / (4 + 5 + 6)
print(answer_a)

# b) 
answer_b = 27 ** (1 / 3)
print(answer_b)

# c) 
answer_c = 3 ** (5 + (3 / 4))
print(answer_c)
````

### ‡∏Ç‡πâ‡∏≠ 10. ‡∏à‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 2 ‡∏Ñ‡πà‡∏≤ ‡πÅ‡∏•‡πâ‡∏ß‡∏™‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á (‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á 2 ‡∏ß‡∏¥‡∏ò‡∏µ) (2 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

````python
# Parallel Assignment
a = float(input("‡∏õ‡πâ‡∏≠‡∏ô‡∏Ñ‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ a: "))
b = float(input("‡∏õ‡πâ‡∏≠‡∏ô‡∏Ñ‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ b: "))

print(f"‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô: a = {a}, b = {b}")


temp = a    
a = b       
b = temp    

print(f"‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà 1 (‡πÉ‡∏ä‡πâ‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ temp): a = {a}, b = {b}")

a, b = b, a 


# -----------------------------------------
a, b = b, a

print(f"‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà 2 (Parallel Assignment): a = {a}, b = {b}")
a = 10
b = 20



print(f"‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà 2: a = {a}, b = {b}")
````
